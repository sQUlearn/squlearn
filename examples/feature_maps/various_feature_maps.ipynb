The example shows various feature maps that are implemented in sQUlearn.

```python
from squlearn.feature_map import *
```

```python
fmap1 = YZ_CX_FeatureMap(num_qubits=5, num_features=5, num_layers=3,c=0.5)
fmap1.draw()
```

```python
fmap2 = QEKFeatureMap(num_qubits=6, num_features=2, num_layers=3)
fmap2.draw()
```

```python
fmap3 = HighDimFeatureMap(num_qubits=10, num_features=45, num_layers=2)
fmap3.draw()
```

```python
fmap4 = ChebyshevTower(num_qubits=6, num_features=2, n_chebyshev=3, num_layers=2)
fmap4.draw()
```

```python
fmap5 = ChebPQC(num_qubits=5, num_features=2, num_layers=2)
fmap5.draw()
```

```python
fmap6 = HZCRxCRyCRz(num_qubits=5, num_features=2, num_layers=2)
fmap6.draw()
```

```python
fmap7 = ChebRx(num_qubits=5, num_features=2, num_layers=2)
fmap7.draw()
```

```python
fmap8 = ZFeatureMap_CX(num_qubits=5, num_features=5, reps=2)
fmap8.draw()
```

```python
fmap9 = QiskitZFeatureMap(num_qubits=5, num_features=5, reps=2)
fmap9.draw()
```

```python
# Feature maps with the same number of qubits and feature dimension can be combined by adding them:
fmap_added = fmap2 + fmap4 + fmap2
fmap_added.draw()
```
