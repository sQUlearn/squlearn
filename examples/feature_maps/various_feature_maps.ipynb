The example shows various feature maps that are implemented in sQUlearn.

```python
from qiskit.circuit.library import TwoLocal
from squlearn.feature_map import *
```

```python
fmap1 = YZ_CX_FeatureMap(num_qubits=5, num_features=5, num_layers=3, c=0.5)
fmap1.draw("mpl")
```

```python
fmap2 = QEKFeatureMap(num_qubits=6, num_features=2, num_layers=3)
fmap2.draw("mpl")
```

```python
fmap3 = HighDimFeatureMap(num_qubits=10, num_features=45, num_layers=2)
fmap3.draw("mpl")
```

```python
fmap4 = ChebyshevTower(num_qubits=6, num_features=2, num_chebyshev=3, num_layers=2)
fmap4.draw("mpl")
```

```python
fmap5 = ChebPQC(num_qubits=5, num_features=2, num_layers=2)
fmap5.draw("mpl")
```

```python
fmap6 = HZCRxCRyCRz(num_qubits=5, num_features=2, num_layers=2)
fmap6.draw("mpl")
```

```python
fmap7 = ChebRx(num_qubits=5, num_features=2, num_layers=2)
fmap7.draw("mpl")
```

```python
fmap8 = ParamZFeatureMap(num_qubits=5, num_features=5, num_layers=2, entangling=True)
fmap8.draw("mpl")
```

```python
"""
This example shows how to wrap a Qiskit QuantumCircuit object into
a sQUlearn FeatureMapBase object. The TwoLocal circuit below can be 
replaced by any other QuantumCircuit
"""
local = TwoLocal(3, "ry", "cx", "linear", reps=2, insert_barriers=True)
fmap9 = QiskitFeatureMap(local)
fmap9.draw("mpl")
```

```python
# Feature maps with the same number of qubits and feature dimension can be combined by adding them:
fmap_added = fmap2 + fmap4 + fmap2
fmap_added.draw("mpl")
```
