The example shows various encoding circuits that are implemented in sQUlearn.

```python
from qiskit.circuit.library import TwoLocal
from squlearn.encoding_circuit import (
    YZ_CX_EncodingCircuit,
    HubregtsenEncodingCircuit,
    HighDimEncodingCircuit,
    ChebyshevPQC,
    ChebyshevRx,
    ChebyshevTower,
    MultiControlEncodingCircuit,
    ParamZFeatureMap,
    QiskitEncodingCircuit,
)
```

```python
enc_circ1 = YZ_CX_EncodingCircuit(num_qubits=5, num_layers=3, c=0.5)
enc_circ1.draw("mpl", num_features=5)
```

```python
enc_circ2 = HubregtsenEncodingCircuit(num_qubits=6, num_layers=3)
enc_circ2.draw("mpl", num_features=2)
```

```python
enc_circ3 = HighDimEncodingCircuit(num_qubits=10, num_layers=2)
enc_circ3.draw("mpl", num_features=45)
```

```python
enc_circ4 = ChebyshevTower(num_qubits=6, num_chebyshev=3, num_layers=2)
enc_circ4.draw("mpl", num_features=2)
```

```python
enc_circ5 = ChebyshevPQC(num_qubits=5, num_layers=2)
enc_circ5.draw("mpl", num_features=2)
```

```python
enc_circ6 = MultiControlEncodingCircuit(num_qubits=5, num_layers=2)
enc_circ6.draw("mpl", num_features=2)
```

```python
enc_circ7 = ChebyshevRx(num_qubits=5, num_layers=2)
enc_circ7.draw("mpl", num_features=2)
```

```python
enc_circ8 = ParamZFeatureMap(num_qubits=5, num_layers=2, entangling=True)
enc_circ8.draw("mpl", num_features=5)
```

```python
"""
This example shows how to wrap a Qiskit QuantumCircuit object into
a sQUlearn EncodingCircuitBase object. The TwoLocal circuit below can be
replaced by any other QuantumCircuit
"""

local = TwoLocal(6, "ry", "cx", "linear", reps=2, insert_barriers=True)
enc_circ9 = QiskitEncodingCircuit(local)
enc_circ9.draw("mpl")
```

```python
# Encoding circuits with the same number of qubits and feature dimension can be combined by adding them:
enc_circ_added = enc_circ2 + enc_circ4 + enc_circ9
enc_circ_added.draw("mpl", num_features=6)
```
